```python
import os

from pydub import AudioSegment, effects
from pydub.playback import play
```

```python
# Load a sound file
test_audio_path = os.path.join(AUDIO_DIR, "and_adele_had_her_own_reasons_for_wanting_to_move_up_here.m4a")
test_audio_data = AudioSegment.from_file(test_audio_path, "m4a")
```

## Volume change

```python
# Play the audio.
play(test_audio_data)
```

```python
# Volume down and play it.
quiter_test_audio_data = test_audio_data - 20
play(quiter_test_audio_data)

print(quiter_test_audio_data.dBFS)
print(quiter_test_audio_data.rms)
```

## Adjust function

```python
def adjust(audio_data, target_dbfs):
    dbfs_diff = audio_data.dBFS - target_dbfs
    return audio_data - dbfs_diff


def adjust_with_normalise(audio_data, target_dbfs):
    normalised_audio_data = effects.normalize(audio_data)
    dbfs_diff = normalised_audio_data.dBFS - target_dbfs
    return normalised_audio_data - dbfs_diff


TARGET_DBFS = -35
```

```python
second_test_audio_path = os.path.join(AUDIO_DIR, "and_it_looks_like_I_am_wearing_blue_gloves.mp3")
second_test_audio_data = AudioSegment.from_file(second_test_audio_path, "mp3")
play(second_test_audio_data)

print(second_test_audio_data.dBFS)
print(second_test_audio_data.rms)
```

```python
adjusted_second_test_audio_data = adjust(second_test_audio_data, TARGET_DBFS)
play(adjusted_second_test_audio_data)

print(adjusted_second_test_audio_data.dBFS)
print(adjusted_second_test_audio_data.rms)
```

```python
adjusted_second_test_audio_data = adjust_with_normalise(second_test_audio_data, TARGET_DBFS)
play(adjusted_second_test_audio_data)

print(adjusted_second_test_audio_data.dBFS)
print(adjusted_second_test_audio_data.rms)
```

```python
adjusted_quiter_test_audio_data = adjust(quiter_test_audio_data, TARGET_DBFS)
play(adjusted_quiter_test_audio_data)

print(adjusted_quiter_test_audio_data.dBFS)
print(adjusted_quiter_test_audio_data.rms)
```

```python
adjusted_quiter_test_audio_data = adjust_with_normalise(quiter_test_audio_data, TARGET_DBFS)
play(adjusted_quiter_test_audio_data)

print(adjusted_quiter_test_audio_data.dBFS)
print(adjusted_quiter_test_audio_data.rms)
```
